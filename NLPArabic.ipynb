```python
def remove_punct(text):
  text_nopunct = "".join([char for char in text if char not in string.punctuation])
  return text_nopunct

fb['BodyText_Arabic_nopunc'] = fb['BodyText_Arabic'].apply(lambda x: remove_punct(x))
fb.tail(10)
```

```python
def tokenize(text):
  tokens = word_tokenize(text)
  return tokens

fb['BodyText_Arabic_tokenized'] = fb['BodyText_Arabic_nopunc'].apply(lambda x: tokenize(x))
fb.head()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('arabic'))
stop_words
```

```python
# Words to keep for semantic analysis in Arabic
words_to_keep_arabic = ["لا", "لكن", "ولا","ولكن","ليسوا", "غير", "لكنما", "لن", "ليسا", "ليست", "ليس",
               "صبر" ,  "لولا"  , "لوما"  ,"صبرا" ,"صدقا" ,"حمدا", "لم","جميع" "ليس", "أبدا", "ترك", "ما"]

# Create a set of custom stop words by excluding words_to_keep from the stop_words set
custom_stop_words_arabic = set(word for word in stop_words if word not in words_to_keep_arabic)

# Define a function to remove custom stop words
def remove_custom_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in custom_stop_words_arabic]
    return text

# Apply the remove_custom_stopwords function to the tokenized text
fb['BodyText_Arabic_nostop'] = fb['BodyText_Arabic_tokenized'].apply(lambda x: remove_custom_stopwords(x))
fb.tail()
```

```python
# Function to stem Arabic text using Snowball stemmer
from snowballstemmer import stemmer

# Assuming you have already initialized the Arabic stemmer (ar_stemmer)
ar_stemmer = stemmer("arabic")

# Function to stem Arabic text using Snowball stemmer
def stem_arabic_text(word_list):
    # Stem each word using Snowball stemmer
    stemmed_words = [ar_stemmer.stemWord(word) for word in word_list]

    return stemmed_words

# Apply the stemming function to the 'BodyText_Arabic_nostop' column in your DataFrame
fb['body_text_Stemmed'] = fb['BodyText_Arabic_nostop'].apply(stem_arabic_text)
# Display the stemmed DataFrame
fb.tail()
```

```python
from collections import Counter
def most_frequent_words(target_value, top_n):
    filtered_fb = fb[fb['label'] == target_value]

    # Combine the lemmatized text from the filtered rows into a single list
    stemmed_text = [word for sublist in filtered_fb['body_text_Stemmed'] for word in sublist]

    # Remove specific characters (e.g., commas) from the text
    cleaned_text = [word.replace('،', '') for word in stemmed_text]

    # Calculate word frequencies
    word_counts = Counter(cleaned_text)

    # Get the top N most frequent words
    top_words = word_counts.most_common(top_n)

    return top_words

top_words_teaching_1 = most_frequent_words( 1, 30)
top_words_teaching_0 = most_frequent_words( 0, 30)
top_words_teaching_minus_1 = most_frequent_words( -1, 30)

print("\n most frequent words in positive feedbacks:")
print(top_words_teaching_1)

print("\n most frequent words in neutral feedbacks:")
print(top_words_teaching_0)

print("\n most frequent words in negative feedbacks:")
print(top_words_teaching_minus_1)
```

```python
# Positive Keywords
positive_keywords = ['جيد', 'في', 'جيدة', 'على', 'التدريس', 'المحاضرات', 'جدًا', 'من', 'الالتزام', 'بالمواعيد',
                      'كل', 'جميع', 'إنه', 'الجامعة', 'التفاعل', 'مع', 'أن', 'الطلاب', 'هو', 'المحاضرة', 'ممتاز',
                      'تسليم', 'طريقة', 'شيء', 'المعلم', 'والمواعيد', 'جيدًا', 'هذه', 'المعلمين']

# Neutral Keywords
neutral_keywords = ['من', 'ولكن', 'التدريس', 'في', 'على', 'جيد', 'بعض', 'التفاعل', 'الطلاب', 'بالمواعيد',
                     'إلى', 'المعلمين', 'لا', 'مع', 'تسليم', 'الالتزام', 'كل', 'متوسط', 'ليس', 'جيدًا', 'أن',
                     'المحاضرين', 'بشكل', 'أيضًا', 'لكن', 'يتم', 'الموضوع', 'ما', 'المحاضرة']

# Negative Keywords
negative_keywords = [ 'من', 'بعض', 'لا', 'ليس', 'جيدة', 'المعلمين', 'الموضوع', 'المحاضرة', 'يتعلق',
                      'المحاضرات', 'حيث', 'ولكن', 'مع', 'جدًا', 'الطلاب', 'هذا', 'بين', 'جيدًا', 'فيما', 'فإنها',
                      'الذين', 'ولا', 'يتفاعلون', 'أنهم', 'جميع', 'أعضاء', 'هيئة', 'التدريس', 'جيدا']


# Function to count positive, neutral, and negative words in a text
def count_sentiment_words(text):
    positive_count = sum(text.count(keyword) for keyword in positive_keywords)
    neutral_count = sum(text.count(keyword) for keyword in neutral_keywords)
    negative_count = sum(text.count(keyword) for keyword in negative_keywords)
    return positive_count, neutral_count, negative_count

# Apply the count_sentiment_words function to each feedback
fb[['positive_count', 'neutral_count', 'negative_count']] = fb['BodyText_Arabic_tokenized'].apply(count_sentiment_words).apply(pd.Series)

# Function to classify sentiment based on keyword counts
def classify_sentiment_with_counts(row):
    if row['positive_count'] > row['neutral_count'] and row['positive_count'] > row['negative_count']:
        return 1  # Positive sentiment
    elif row['negative_count'] > row['neutral_count'] and row['negative_count'] > row['positive_count']:
        return -1  # Negative sentiment
    else:
        return 0  # Neutral sentiment

# Apply the sentiment classification function to each row
fb['predicted_sentiment_with_counts'] = fb.apply(classify_sentiment_with_counts, axis=1)

# Calculate accuracy
accuracy = (fb['predicted_sentiment_with_counts'] == fb['label']).sum() / len(fb)

print(f"Accuracy: {accuracy * 100:.2f}%")
```
